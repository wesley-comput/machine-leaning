```python
#feature
#pelo longo?
#perna curta?
#faz auau?

porco1 = [0, 1, 0]
porco2 = [0, 1, 1]
porco3 = [1, 1, 0]

cachorro1 = [0, 1, 1]
cachorro2 = [1, 0, 1]
cachorro3 = [1, 1, 1]

treino_x = [porco1, porco2, porco3, cachorro1, cachorro2, cachorro3]
treino_y = [1, 1, 1, 0, 0, 0] #labels
```

## Treinando o modelo

```python
from sklearn.svm import LinearSVC

modelo = LinearSVC()
modelo.fit(treino_x, treino_y)
```

## Prevendo um resultado aleatório

```python
animal_misterioso = [1, 1, 1]
modelo.predict([animal_misterioso])
```

## Preparando o modelo com vários elementos

```python
misterio1 = [1, 1, 1]
misterio2 = [1, 1, 0]
misterio3 = [0, 1, 1]

teste_x = [misterio1, misterio2, misterio3]
teste_y = [0, 1, 1]
```

## Prevendo o resultado

```python
previsoes = modelo.predict(teste_x)
print(previsoes)
```

```python
previsoes == teste_y
```

```python
corretos = (previsoes == teste_y).sum()
total = len(teste_x)
taxa_acerto = (corretos / total) * 100
print("Taxa de acerto: %.2f ", taxa_acerto)
```

```python
from sklearn.metrics import accuracy_score

taxa_acerto = accuracy_score(teste_y, previsoes)
print("Taxa de acerto: %.2f ", taxa_acerto * 100)
```
